```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 256
x = np.linspace(0, np.pi*2, num=D, endpoint=False)
```

```python
data = np.tile(np.sin(x), (D,1))
```

```python
plt.axis('off')
img = plt.imshow(data, cmap='gray')
# plt.savefig('grating2.png',bbox_inches='tight',transparent=True, pad_inches=0)
```

```python
from PIL import Image as im
im_data = im.fromarray(((data + 1)*(2**15-1)).astype(np.uint16))
im_data.save('grating.png')
```

```python
data2  = np.expand_dims(np.sin(x) >= 0,1) ^ np.expand_dims(np.sin(x) >= 0,1).T
print(data2.shape)
plt.axis('off')
img = plt.imshow(data2, cmap='gray')
im_data = im.fromarray(((data2 + 1)*(2**15-1)).astype(np.uint16))
im_data.save('checkerboard.png')
```

```python
data3 = np.random.randn(D,D)
plt.axis('off')
img = plt.imshow(data3, cmap='gray')
im_data = im.fromarray(((data3 + 1)*(2**15-1)).astype(np.uint16))
im_data.save('whitenoise.png')
```

```python
g = np.expand_dims(np.exp(-(x - np.pi)**2/2),1)
data4 = (1 - g * g.T)
data4 = np.swapaxes(np.tile(data4, (4,1,1)),0,2) # generate RGBA!
data4[:,:,3] = (1 - data4[:,:,3])
data4[:,:,:3] = 0.1*data4[:,:,:3]
plt.axis('off')
img = plt.imshow(data4, cmap='gray')
im_data = im.fromarray(((data4)*(2**8-1)).astype(np.uint8))
im_data.save('gaussian.png')
```

```python
g = np.expand_dims(np.sin(x/2),1)
data4 = (1 - g * g.T)
data4 = np.swapaxes(np.tile(data4, (4,1,1)),0,2) # generate RGBA!
data4[:,:,3] = (1 - data4[:,:,3])
# data4[:,:,:3] = data4[:,:,:3]
plt.axis('off')
img = plt.imshow(data4, cmap='gray')
im_data = im.fromarray(((data4)*(2**8-1)).astype(np.uint8))
im_data.save('raised_sine.png')
```
